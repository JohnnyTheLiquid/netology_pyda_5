Задание 1
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head(3)
```

```python
""" Выделяем список пользователей, поставивших 100 и более оценок """
rating_grouped_count = ratings.groupby('userId').count().reset_index()
list100 = (rating_grouped_count [ rating_grouped_count['rating'] >= 100]['userId']).tolist()

list100
```

```python
ratings_group_max = ratings.groupby('userId').max().reset_index()
ratings_group_max.head(3)
```

```python
ratings_group_min = ratings.groupby('userId').min().reset_index()
ratings_group_min.head(3)
```

```python
ratings_group = ratings_group_min.merge(ratings_group_max, how = 'left', on = 'userId' )
""" Можно было бы оставить и по умолчанию inner, резульат тот же. Но left возможно выявил бы пробелы в предыдущих стадиях, 
хотя им неоткуда взяться - мы делим весь датафрейм на мин и макс, в худшем случае они равны. Поле timestamp присваивается 
автоматически, NaN, False и т.п. по нему быть не должно"""

ratings_group[ratings_group ['userId'].isin(list100)].head()
```

```python
""" Находим требуемое среднее значение разницы макс и мин timestamp для пользователей из списка list100"""
(
(ratings_group[ratings_group ['userId'].isin(list100)])['timestamp_y'] 
- 
(ratings_group[ratings_group ['userId'].isin(list100)])['timestamp_x']
).mean()
```

Задание 2
Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
""" Чтобы использовать join делаем индексами client_id """
rzd_n = rzd.set_index('client_id')
air_n = air.set_index('client_id')
auto_n = auto.set_index('client_id')
```

```python
(rzd_n.join(air_n, how = 'outer')).join(auto_n,how = 'outer')
```

```python
# Произодим замену NaN на 0 
all_transp = (rzd_n.join(air_n, how = 'outer')).join(auto_n,how = 'outer').fillna(0)
all_transp
```

```python
all_transp.info()
```

```python
""" Вторая часть задания - объединить с таблицей client_base; чтобы итоговая таблица выглядела более логично,
привязываем all_transp к client_base, а не наоборот"""
client_base.merge(all_transp, how = 'left', on = 'client_id')
```

```python
""" Для проверки, что ничего не потярялось можно вызвать метод .info() и сверить количество строк"""
(client_base.merge(all_transp, how = 'left', on = 'client_id')).info()
```

Задание 3
В задаче сквозной аналитики вам предоставили данные по местоположению пользователей. Т. е. для каждого user_id известна последовательность координат (широта/долгота), когда они требовались приложению для полноценной работы. Как бы вы добавили эти сведения в таблицу визитов и покупок?

```python
visits = pd.DataFrame(
    {
        'user_id': [11, 22, 55, 11, 77],
        'source': ['ad', 'yandex', 'email', 'google', 'ad']
    }
)

visits = visits[['user_id', 'source']]
visits
```

```python
purchases = pd.DataFrame(
    {
        'user_id': [11, 22, 55, 11, 99],
        'category': ['Спорт', 'Авто', 'Дача', 'Спорт', 'Авто'],
    }
)

purchases = purchases[['user_id', 'category']]
purchases
```

```python
geo = pd.DataFrame(
{
    'user_id' : [11,22,55,77,99],
    'coordinates' : [(55.619387, 37.546973),(55.838701, 37.755713), (55.756700, 37.430243), (55.592955, 37.687049), (55.772959, 37.780432)]
})
geo
```

```python
visits_purchases = visits.merge(purchases, how = 'outer', on = 'user_id')
visits_purchases
```

```python
visits_purchases.drop_duplicates(keep = 'first', inplace = True)
visits_purchases
```

```python
visits_purchases_geo = visits_purchases.merge(geo, how='left', on = 'user_id')
""" В итоге, получаем таблицу со всеми данными, из которой можно уже отфильтровывать, сводить, группировать """
visits_purchases_geo
```

```python
purch_pivot_1 = visits_purchases_geo.pivot_table(index ='user_id', columns = 'category', 
                                               values = 'source', aggfunc = 'size', fill_value = 0)
""" Аналогично тому, как делали на занятии - срез пользователь-категория """
purch_pivot_1
```

```python
purch_pivot_2 = visits_purchases_geo.pivot_table(index ='coordinates', columns = 'source', 
                                               values = 'category', aggfunc = 'size', fill_value = 0)
""" Связь геолокации и канала, в случае, когда покупка была без канала - ожидаемо, привязка в таблице не выводится"""
purch_pivot_2
```
