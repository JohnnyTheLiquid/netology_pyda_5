Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. 
Определите какому фильму было выставлено больше всего оценок 5.0.

Решение выполняем ступенчато. Выясняем структуры таблиц (колонки), собираем из двух таблиц одну, чтобы получить соответствие названий и рейтингов. Отфильтровываем те, у которых рейтинг 5 и группируем 'по movieId'. К получившейся таблице добавляем колонку с подсчетом 'userId', то есть, сколько пользователей выставили рейтинг (уже отфильрован только на 5). Выводим итоговый результат сортировкой и выборкой 1 строки.

```python
import pandas as pd

rat=pd.read_csv('C:/Users/User/Documents/My Python Netology/ДЗ 6/ml-latest-small/ml-latest-small/ratings.csv')

rat.head(3)
```

```python
mov=pd.read_csv('C:/Users/User/Documents/My Python Netology/ДЗ 6/ml-latest-small/ml-latest-small/movies.csv')

mov.head(3)
```

```python
mov_rat = mov.merge(rat, on = 'movieId', how = 'left')

mov_rat.head(3)
```

```python
mov_rat[mov_rat['rating'] == 5].head(3)
```

```python
mov_rat_5 = mov_rat[mov_rat['rating'] == 5].groupby('movieId').head(10)

mov_rat_5
```

```python
mov_rat_5['count'] = mov_rat_5['userId'].count()

mov_rat_5.sort_values(by = 'count', ascending = False).head(1)
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) 
категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

Решение выполянем ступенчато. Ищем, как называются страны и создаем выборку по этим странам. Далее выбираем только с нужными категориями. Добавляем остальные условия, выводим только колонку 'quantity' с методом sum()

```python
import pandas as pd

power_data = pd.read_csv('power.csv')

power_data.head(5)
```

```python
power_data[ power_data['country'].str.contains('nia', case=False) ]['country'].unique()
```

```python
power_data[ power_data['country'].str.contains('via', case=False) ]['country'].unique()
```

```python
pw_baltic = power_data[ 
(power_data['country'] == 'Estonia') | 
(power_data['country'] == 'Lithuania') | 
(power_data['country'] == 'Latvia') 
]

pw_baltic.head(5)
```

```python
pw_baltic_cat_4_12_21 = pw_baltic[ 
(pw_baltic['category'] == 4) | 
(pw_baltic['category'] == 12) | 
(pw_baltic['category'] == 21) 
 ]

pw_baltic_cat_4_12_21.head(5)
```

```python
result = pw_baltic_cat_4_12_21['quantity'][
(pw_baltic_cat_4_12_21['quantity'] > 0) &
(pw_baltic_cat_4_12_21['year'] > 2004) &
(pw_baltic_cat_4_12_21['year'] < 2011)
].sum()

result
```
