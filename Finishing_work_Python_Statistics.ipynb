#### 5.С использованием t-test статистики проверить взаимосвязи между несколькими переменными.

Т-тест в нашем случаем может проверить либо две независимые выборки, либо выборку по отношению к некоему заданному среднему.

```python
"""Проверим гипотезу выгорания от нагрузки, то есть, насколько нагрузка в часах у уволившихся отличалась от средней нагрузки.
Нулевая гипотеза - средние нагрузки в выборке и в генеральной совокупности не отличаются. В этом случае выборки не являются
независимыми и мы проверяем выборку и медиану генеральной совокупности"""

left = df[df['left'] == 1]['average_montly_hours']
stats.ttest_1samp(left, df['average_montly_hours'].mean())
```

**pvalue** получилось крайне малым, то есть нулевую гипотезу принять нельзя. Соответственно, по данному тесту возможна связь между повышенной нагрузкой и фактом увольнения сотрудника.

```python
"""Проверим ту же гипотезу, сравнив выборки уволившихся и оставшихся"""
staying = df[df['left'] != 1]['average_montly_hours']
print(len(left)), print(len(staying))
print(statistics.variance(left)), 
print(statistics.variance(staying))
```

```python
"""Мы видим, что выборки достаточно большие и содержат разное количество значений. Чтобы сравнить их, выберем сэмплы"""
left_samp = left.sample(1000)
staying_samp = staying.sample(1000)
print(statistics.variance(left_samp)) 
print(statistics.variance(staying_samp))
```

```python
"""Поскольку отклонения сильно отличаются, проводим Т-тест c критерием Уелча"""
stats.ttest_ind(left_samp, staying_samp, equal_var=False)
```

**p-value** получилось существенно ниже порогового значения 0.05, что говорит о том, что нулевая гипотеза о сходстве выборок не может быть принята. Выборки не могут быть признаны статистически не отличающимися. То есть, гипотеза о выгорании не может быть отвергнута.

**Примечание** - если взять сэмпл меньшего размера, например, 100, что в соответствии с трактовками ЦПТ может считаться   достаточным, то результат получается обратным, p-value будет около 0.17, что существенно выше порогового значения 0.05. Практический пример того, что размер выборки влияет.

```python
"""Визуализируем распределения выборки и генеральной совокупности, применив distplot. 
Результаты обсудим далее, в задании 10"""

sns.distplot(left)
sns.distplot(df['average_montly_hours'])
plt.title('Сравение плотности вероятности количества часов уволившихся и генеральной совокупности')
plt.legend(['уволившиеся', 'генеральная совокупность'])
plt.show()
```

```python
"""Проверим, есть ли связь уровня зарплаты и времени работы в компании. 
Ожидаемо, из трех уровней зарплат, самый массовый - low"""

df['salary'].describe()
```

```python
"""Нулевая гипотеза - выборка с низким уровнем зарплаты и медиана генеральной совокупности - не имеют отличий по времени работы
в компании."""

low_salary = df[df['salary'] == 'low']['time_spend_company']
stats.ttest_1samp(low_salary, df['time_spend_company'].mean())
```
