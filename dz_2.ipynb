# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
rng = np.random.RandomState(123)

result_list_1 = []

for flip in range(10):
  result_list_1.append(rng.randint(1,7))
  
print(result_list_1)

sns.distplot(result_list_1, bins=6)
```

Напишите код для моделирования 1000 подбрасываний игрального кубика. Начертите диаграму histogram. Почему график изменился? Как применить центральную предельную теорему?

```python
result_list_2 = []

for flip in range(1000):
  result_list_2.append(rng.randint(1,7))
  
print(result_list_2)

sns.distplot(result_list_2, bins=6)
```

ЦПТ имеет дело с распределением суммы случайных величин. Как мы видим, при увеличении числа испытаний, распределение количества выпадений каждого из значений приближается к равномерному, что ожидаемо. Результат каждого следующего броска не зависит от предыдущего, то есть, при достаточно большом числе бросков, каждая сторона выпадет примерно одинаковое количество раз.
Но, если провести, например, 1000 испытаний в каждом из которых считать среднее от 10 бросков игральной кости, то такое распределение будет приближаться к нормальному с пиком в точке математического ожидания, а именно, 3.5

```python
sum_list = []
result_list = []

for series in range(1000):
  for flip in range(10):
    result_list.append(rng.randint(1,7))
  sum_list.append(sum(result_list) / len(result_list))
  

sns.distplot(sum_list)
```
